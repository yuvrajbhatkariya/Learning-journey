```python
import spacy 
nlp = spacy.load("en_core_web_lg")
```

```python
doc = nlp("hey yuvraj")
doc.vector
```

```python
test_token = nlp("yuvraj")
doc = nlp("person king poor hero corrupt queen happy")
for token in doc:
    print(f"{token.text} <-> {test_token.text} : ",token.similarity(test_token))
```

```python
import pandas as pd 
df = pd.read_csv("IMDB Dataset.csv")
df.head()
```

```python
df["num_review"] = df["review"].apply(lambda x : nlp(x).vector)
df.head()
```

```python
df["num_sentiment"] = df["sentiment"].map({
    "positive" : 0,
    "negative" : 1
})
df.head()
```

```python
df.to_csv("Embed_IMDB_data.csv",index=False)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df.num_review, df.num_sentiment , test_size=0.20, random_state = 2026)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
```

```python
import numpy as np
x_train_v = np.stack(x_train)
x_test_v = np.stack(x_test)
```

```python
x_train_v.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scalar = MinMaxScaler()
x_train_scaled = scalar.fit_transform(x_train_v)
x_test_scaled = scalar.fit_transform(x_test_v)
```

```python
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB(alpha=1)
nb.fit(x_train_scaled,y_train)
```

```python
from sklearn.metrics import classification_report,accuracy_score
y_pred = nb.predict(x_test_scaled)
print(f"Acuracy : {accuracy_score(y_test,y_pred)}")
print(classification_report(y_test,y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5,metric="euclidean")
knn.fit(x_train_v,y_train)
```

```python
y_pred = knn.predict(x_test_v)
print(f"Acuracy : {accuracy_score(y_test,y_pred)}")
print(classification_report(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(
    n_estimators=200,
    max_depth=None,
    random_state=42,
    n_jobs=-1
)
rf.fit(x_train_v,y_train)
```

```python
y_pred = rf.predict(x_test_v)
print(f"Acuracy : {accuracy_score(y_test,y_pred)}")
print(classification_report(y_test,y_pred))
```

```python
from catboost import CatBoostClassifier
cat = CatBoostClassifier(
    iterations=500,
    learning_rate=0.1,
    depth=6,
    loss_function="Logloss",   
    eval_metric="Accuracy",
    random_seed=42,
    verbose=100
)
cat.fit(x_train_v,y_train)
```

```python
y_pred = cat.predict(x_test_v)
print(f"Acuracy : {accuracy_score(y_test,y_pred)}")
print(classification_report(y_test,y_pred))
```

```python
from xgboost import XGBClassifier
xgb = XGBClassifier(
    n_estimators=500,
    max_depth=6,
    learning_rate=0.1,
    subsample=0.8,
    colsample_bytree=0.8,
    objective="binary:logistic",
    eval_metric="logloss",
    random_state=42,
    n_jobs=-1
)
xgb.fit(x_train_v,y_train)
```

```python
y_pred = xgb.predict(x_test_v)
print(f"Acuracy : {accuracy_score(y_test,y_pred)}")
print(classification_report(y_test,y_pred))
```

```python
def testing(text):
    doc = nlp(text)
    vec = doc.vector
    v_2d = vec.reshape(1, -1)
    y = xgb.predict(v_2d)
    if(y==0):
        print("Positive")
    else:
        print("negative")

testing("Good performance, but the build quality is poor.")
    
```
