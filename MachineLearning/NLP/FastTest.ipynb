```python
import re
text = "I'm restless and restless, it's been a month now, boy. What do you mean?"
clean_text = re.sub(r"[^\w\s]", "", text)
clean_text
```

```python
def prepros(text):
    clean_text = re.sub(r"[^\w\s]", "", text)
    clean_text = re.sub(r"[ \n]+"," ",clean_text)
    return clean_text.strip().lower()

text = "I'm restless and restless, it's been a month now, boy. What do you mean?"
prepros(text)
```

### **Build my own model: -**

```python
df["prepros_text"] = df["text"].apply(lambda x : prepros(x))
df.sample(7)
```

```python
df.to_csv("test.text",columns=["prepros_text"],index=False,header=None)
```
