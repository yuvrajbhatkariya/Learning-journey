```python
import numpy as np

VECTOR_SIZE = wv.vector_size

def preprocess(text):
    if not isinstance(text, str) or text.strip() == "":
        return np.zeros(VECTOR_SIZE)

    doc = nlp(text)
    filtered_tokens = []

    for token in doc:
        if token.is_stop or token.is_punct:
            continue
        if token.lemma_ in wv:
            filtered_tokens.append(token.lemma_)

    if len(filtered_tokens) == 0:
        return np.zeros(VECTOR_SIZE)

    return wv.get_mean_vector(filtered_tokens)
```

```python
preprocess("What the fuck are you doing ????")
```

```python
df["num_text"] = df["text"].apply(lambda x: preprocess(x))
df.sample(8)
```

```python
df.to_csv("Email_preprocess.csv", index=False)
```
