### A blank pipeline means it only has the tokenizer (splits text into words, punctuation, etc.)

```python
nlp = spacy.blank("en")   # object

doc  = nlp("Dr. Strange is a Doctor. Is Dr. strange belives in magic?")

for token in doc: 
    print(token)
```

```python
doc  = nlp('''Tony Stark booked tickets in Los Angeles for Avengers, sending details to bruce.    banner@avengers.com
. Natasha called him on +1-310-555-7890, reminding about the $1200 budget for Spider-Man’s premiere with Doctor Strange.''')
```

```python
for token in doc:
    print(token," | ",token.pos_," | ",token.lemma_)

# POS - part of speech (noun ,pronoun, adjective, verb)
# lemmitization
```

```python
para = nlp('''Emma Johnson traveled to Paris for a conference, where she emailed contact@globalmeet.com
. She later called +44-20-7946-0958 to confirm the hotel booking worth €750 arranged by Michael on monday 12/08/2025 .''')

for ent in para.ents:
    print(ent.text, " | ", ent.label_ ," | ", spacy.explain(ent.label_))
```

```python
from spacy import displacy

displacy.render(para,style="ent")
```

## **Special case like to words are mix : -**

```python
doc = nlp('''Aman Verma giveme a email of Priya that is  priya.shah@example.com.He want me to ask her about the timming of meeting on 12/09/2025.''')

from spacy.symbols import ORTH

nlp.tokenizer.add_special_case("giveme",[
    {ORTH : "give"},
    {ORTH : "me"}
])

token = [token.text for token in doc]
token
```

```python
for sentence in doc.sents:
    print(sentence)
```
