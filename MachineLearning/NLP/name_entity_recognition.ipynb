```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
nlp.pipe_names
```

```python
doc = nlp("Apple is looking at buying U.K. startup for $1 billion. Tesla Inc is going to acquire Twitter Inc for $44 billion.")
for ent in doc.ents:
    print(ent.text ,"|", ent.label_,"|",spacy.explain(ent.label_))
```

```python
from spacy import displacy
displacy.render(doc,style="ent")
```

```python
doc1 = nlp("Apple is looking at buying U.K. startup for $1 billion. Tesla is going to acquire Twitter for $44 billion.")
```

```python
from spacy.tokens import Span

s1 = Span(doc1,12,13,label= "ORG")
s2 = Span(doc1,17,18,label= "ORG")

doc1.set_ents([s1,s2],default="unmodified")
```

```python
for ent in doc1.ents:
    print(ent.text ,"|", ent.label_)
```

```python
from spacy import displacy
displacy.render(doc1,style="ent")
```
