### Preprocessing : -

```python
import re

def clean_text(text):

    text = str(text).lower()   # convert to string first
    
    text = re.sub(r'@\w+', '', text)        # remove mentions
    text = re.sub(r'#\w+', '', text)        # remove hashtags
    text = re.sub(r'\b\d{1,2}:\d{2}(\s?[ap]\.?m\.?)?', '', text)  # remove time
    text = re.sub(r'\d+', '', text)         # remove numbers
    text = re.sub(r'[^\w\s]', '', text)     # remove punctuation
    text = re.sub(r'\s+', ' ', text).strip()

    return text
```

```python
df1['new_title'] = df1['title'].apply(clean_text)
df1.head()
```
