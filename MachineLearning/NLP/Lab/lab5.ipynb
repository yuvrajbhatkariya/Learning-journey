### **Here we remoing Puntuation : -**

```python
import re
def removing_punctuation(text):
    clean = re.sub(r'[^\w\s]', '', text)
    return clean

print(removing_punctuation("hey buddy ! how is going &&&& & / ?"))
```

```python
def num_asssign(tokens):
    word_to_index = {}
    index = 1

    for word in tokens:
        if word not in word_to_index:
            word_to_index[word] = index
            index = index + 1

    return word_to_index

# token = n_gram("Hey , i am here . where are you ?",6)
# assign = num_asssign(token)
# print(assign)
```

### **Main Function : -**

```python
def n_gram(text, n):
    text = text.lower()
    text = removing_punctuation(text)
    tokens = text.split()

    final_dict = {}
    counter = 1  

    j = 1
    while (j <= n):
        n_tokens = [" ".join(tokens[i:i+j]) for i in range(len(tokens)-j+1)]
        
        for word in n_tokens:
            final_dict[word] = counter
            counter += 1
        
        j += 1

    print(final_dict)
```
