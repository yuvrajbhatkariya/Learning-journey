### **Use the concept in movie review classification: -**

```python
import spacy
import pandas as pd

# Load spaCy with heavy components disabled
nlp = spacy.load(
    "en_core_web_sm",
    disable=["parser", "ner", "tagger"]
)
```

```python
def remove_stopwords(doc):
    return " ".join(
        token.text
        for token in doc
        if not token.is_stop
        and not token.is_punct
        and (token.is_alpha or token.is_digit or token.like_num)
    )
```

```python
# Load dataset
df = pd.read_csv("IMDB Dataset.csv")
df.head()
```

```python
len(df["review"][0])
```

```python


# Process text in batches
docs = nlp.pipe(
    df["review"],
    batch_size=1000
)

# Apply cleaning
df["noStop_words"] = [remove_stopwords(doc) for doc in docs]
```
