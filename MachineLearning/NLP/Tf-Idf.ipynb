### **Apply on emotion dataset to find the emotion in data:-**

```python
import pandas as pd

df = pd.read_csv("emotion_dataset/train.txt", header=None,names=["text"])
df.head()
```

```python
df.info()
```

```python
df[["text","label"]] = df["text"].str.split(";",expand=True)
df.head()
```

```python
df.info()
```

```python
df.label.value_counts()
```

```python
df["label_num"] = df["label"].map({
    "joy": 0,
    "sadness": 1,
    "anger": 2,
    "fear": 3,
    "love": 4,
    "surprise" : 5
})
df.head()
```

```python
NEGATIONS = {
    "no",
    "not",
    "never",
    "none",
    "nothing",
    "nowhere",
    "neither",
    "nor",
    "without"
}
def prepros(doc):
    tokens = []

    negate_count = 0 

    for token in doc:

        if token.is_punct:
            negate_count = 0
            continue

        if token.dep_ == "neg" or token.text.lower() in NEGATIONS:
            tokens.append("not")
            negate_count = 1
            continue

        if token.is_stop:
            continue

        lemma = token.lemma_.lower()

        if negate_count > 0:
            lemma = "not_" + lemma
            negate_count -= 1

        tokens.append(lemma)

    return " ".join(tokens)
```

```python
df = pd.read_csv("prepros_emotion.csv")
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train ,y_test =train_test_split(df.prepros_text,df.label_num,test_size=0.20,stratify=df.label_num,random_state=2026)
print(x_train.shape)
print(y_test.shape)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.naive_bayes import MultinomialNB
pipeline = Pipeline([
    ("tfidf",TfidfVectorizer(ngram_range=(1,2))),
    ("nb",MultinomialNB(alpha=0.1))
])
```

```python
pipeline.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report

y_pred = pipeline.predict(x_test)

print("Accuracy:",accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
RF = Pipeline([
    ("tfidf",TfidfVectorizer(
        ngram_range=(1,2),
        max_features=3000)),
    ("rf",(RandomForestClassifier(
        n_estimators=200,
        max_depth=None,
        random_state=42,
        n_jobs=-1
    )))
])

RF.fit(x_train,y_train)
y_pred = RF.predict(x_test)

print("Accuracy:",accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

### **Testing Part : -**

```python
reverse_mapping = {
    0: 'joy',
    1: 'sadness',
    2: 'anger',
    3: 'fear',
    4: 'love',
    5: 'surprise'
}

```

```python
def testing(text):
    doc = nlp(text)
    prepros_text = prepros(doc)
    ans = pipeline.predict([prepros_text])[0]
    emotion = reverse_mapping[ans]
    return emotion

text = "I can’t stop smiling today because everything is finally going my way."
testing(text)
```

```python
text = "I am not angry, just disappointed."
testing(text)
```

```python
tests = [
    "I feel very happy and satisfied with my life right now.",
    "I am not happy with how things turned out.",
    "I am extremely angry about how unfair this situation is.",
    "I am scared that something bad might happen to me.",
    "I truly care about you and feel deeply connected to you.",
    "I never expected this news, it completely shocked me."
]

for t in tests:
    print(t, "->", testing(t))
```
