### **Stemming using nltk:-**

```python
from nltk.stem import PorterStemmer,SnowballStemmer
stemmer = PorterStemmer()
snow_stemmer = SnowballStemmer("english")
```

```python
words = ["studies", "studying", "studied","fishes","ate","beautiful","happily","better","went","feet","best"]
for word in words:
    print(word,"|",stemmer.stem(word))
```

```python
words = ["studies", "studying", "studied","fishes","ate","beautiful","happily","better","went","feet","best"]
for word in words:
    print(word,"|",snow_stemmer.stem(word))
```

### **Lemmatization using Spacy :-**

```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
doc = nlp("I am going to eat Fried Rice.I am eating too fast that it will took me just 5 min to eat 1 plate rice.")

for token in doc:
    print(token, "|",token.lemma_)
```

```python
nlp.pipe_names
```

### **Custom Lemmatization using attribute_ruler**

```python
custom = nlp.get_pipe("attribute_ruler")

custom.add([[{"TEXT":"bro"}],[{"TEXT":"brah"}]],{"LEMMA":"brother"})


doc = nlp("Hey! bro i am hungy,brah don't ignore me,I am your Brother")

for token in doc:
    print(token,"|", token.lemma_)
```
