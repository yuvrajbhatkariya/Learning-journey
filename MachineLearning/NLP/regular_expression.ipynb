```python
import re
```

```python
text =  {1 : "You can reach me at johndoe@example.com or call me at +91 7470713290.",
2 : "Please contact support via support123@mailservice.org, phone: (415) 555-2671.",
3 : "For inquiries, email: priyasharma99@gmail.com, mobile: 9876543210.",
4 : "Customer care: helpdesk@company.co.uk, contact no: +44 7700 900123.",
5 : "Drop a mail to rajiv.kumar@university.edu or dial 022-27654321.",
}
```

```python
phone_pattern = '\d{10}|\+*\d{2}\s\d{10}|\+*\d{2}\s\d{4}\s\d{6}|\(\d{3}\)\s\d{3}\-\d{4}|\d{3}\-\d{8}'

email_pattern = '[a-zA-Z0-9_]*\@[a-zA-Z]*\.[a-zA-Z]*\.[a-zA-Z]*|[a-zA-Z0-9]*\@[a-zA-Z]*\.[a-z]*'

# email_matches = re.findall(phone_pattern,text5)
# phone_matches = re.findall(email_pattern,text5)
```

```python
for key,value in text.items(): 
    phone_matches = re.findall(phone_pattern,value)
    print(phone_matches)
```

```python
for key,value in text.items(): 
    email_matches = re.findall(phone_pattern,value)
    print(email_matches)
```

```python
print(email_matches)
print(phone_matches)
```
