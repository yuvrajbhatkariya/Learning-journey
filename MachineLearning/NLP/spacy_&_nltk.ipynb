```python
import spacy
nlp = spacy.load("en_core_web_sm")

doc = nlp("Dr. Strange is a Doctor. Is Dr. strange belives in magic?")


for sentence in doc.sents:
    print(sentence)
```

```python
for sentence in doc.sents:
    for word in sentence:
        print(word)
```
