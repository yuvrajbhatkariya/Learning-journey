```python
x_train = x_train/255
print(np.unique(x_train))
x_test = x_test/255
```

### **ANN: -**

```python
import tensorflow as tf
from tensorflow.keras import layers, models
```

```python
x_train.shape
```

```python
ann = models.Sequential([
    layers.Flatten(input_shape = (28,28)),
    layers.Dense(6000,activation = 'relu'),
    layers.Dense(3000,activation = 'relu'),
    layers.Dense(10,activation = 'softmax')
])

ann.compile(optimizer = 'adam',loss = 'sparse_categorical_crossentropy',metrics = ['accuracy'])
ann.fit(x_train,y_train,epochs = 1)
```

```python
y_pred = ann.predict(x_test)
y_classes = [np.argmax(element) for element in y_pred]
y_classes[:5]
```

```python
y_test[:5]
```

```python
from sklearn.metrics import confusion_matrix,classification_report
report = classification_report(y_test, y_classes)

print("Classification Report:\n")
print(report)
```

### **build a convolutional neural network:-**

```python
cnn = models.Sequential([
    layers.Reshape((28, 28, 1), input_shape=(28, 28)),

    layers.Conv2D(filters = 32,kernel_size=(3,3),strides = (1,1),padding = 'same',data_format = None, activation = 'relu'),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(filters = 64,kernel_size = (3,3),padding = 'valid', data_format = None,activation = 'relu'),
    layers.MaxPooling2D((2,2)),

    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

cnn.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

cnn.fit(x_train, y_train, epochs=1)
```

```python
cnn.summary()
```

```python
from sklearn.metrics import confusion_matrix,classification_report

y_pred = cnn.predict(x_test)
y_classes = [np.argmax(element) for element in y_pred]
report = classification_report(y_test, y_classes)

print("Classification Report:\n")
print(report)
```
