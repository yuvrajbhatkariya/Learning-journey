# **Tensorflow Pipeline: -**

```python
import tensorflow as tf
```

```python
stocks_profit = [12,14,5,-1,82,110,-20]
```

### 1. Convert array into dataset

```python

tf_dataset = tf.data.Dataset.from_tensor_slices(stocks_profit)
tf_dataset
```

```python
for profit in tf_dataset:
    print(profit.numpy())
```

```python
for profit in tf_dataset.as_numpy_iterator():
    print(profit)
```

### 2. Apply a filter (to remove negative values) : -

```python
tf_dataset = tf_dataset.filter(lambda x: x>0)
for profit in tf_dataset.as_numpy_iterator():
    print(profit)
```

```python
tf_dataset = tf_dataset.map(lambda x: x*20)
for profit in tf_dataset.as_numpy_iterator():
    print(profit)
```

```python
tf_dataset = tf_dataset.shuffle(20)
for profit in tf_dataset.as_numpy_iterator():
    print(profit)
```

```python
for profit in tf_dataset.batch(3):
    print(profit.numpy())
```

## **All code in one line : -**

```python
tf_dataset = tf.data.Dataset.from_tensor_slices(stocks_profit)

for profit in tf_dataset:
    print(profit.numpy())
```

```python
tf_dataset = tf_dataset.filter(lambda x: x>0).map(lambda y: y*20).shuffle(20).batch(3)
for profit in tf_dataset:
    print(profit.numpy())
```
