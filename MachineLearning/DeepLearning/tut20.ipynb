# **Dropout Regularization in deep Neural networks:-**
###  To prevent from overfitting by removing some neuron in hidden layers.

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
y = df['R']
x = df.drop('R',axis = 1)
```

```python
y = pd.get_dummies(y,drop_first = True)
y.sample(5)
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train, y_test = train_test_split(x,y,test_size = 0.25,random_state=42)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

### **Without dropout regularisation:-**

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Dense(60,input_shape = (60,),activation = 'relu'),
    keras.layers.Dense(30,activation='relu'),
    keras.layers.Dense(15,activation='relu'),
    keras.layers.Dense(1,activation = 'sigmoid')
])
model.compile(loss = 'binary_crossentropy',optimizer='adam',metrics = ['accuracy'])
model.fit(x_train,y_train,epochs= 400,batch_size = 8)
```

```python
model.evaluate(x_test,y_test)
```

### **use Dropout Regularization: -**

```python
model1 = keras.Sequential([
    keras.layers.Dense(60,input_shape = (60,),activation = 'relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(30,activation = 'relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(15,activation = 'relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense( 1,activation = 'sigmoid')
])

model1.compile(loss = 'binary_crossentropy',optimizer='adam',metrics = ['accuracy'])
model1.fit(x_train,y_train,epochs = 100,batch_size = 8)
```

```python
model1.evaluate(x_test,y_test)
```

```python
model.evaluate(x_test,y_test)
```

```python
y_pred = model.predict(x_test)
y_pred1 = model1.predict(x_test)
```

```python
print(y_pred[:10])
```

```python
y_pred = np.round(y_pred)
y_pred1 = np.round(y_pred1)
print(y_pred[:10])
print(y_pred1[:10])
```

```python
from sklearn.metrics import confusion_matrix ,classification_report
print(classification_report(y_test,y_pred))

print("=====================================================================================")
print(classification_report(y_test,y_pred1))
```
