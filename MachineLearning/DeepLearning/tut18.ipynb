### **Convert string data into numeric data: -**

```python
df['TotalCharges'] = df['TotalCharges'].replace(" ", np.nan)  # Replace empty strings with NaN
df.isnull().sum()
```

# **Encoding: -**

```python
def print_unique_col_values(df):
       for column in df:
            if df[column].dtypes=='object':
                print(f'{column}: {df[column].unique()}') 

print_unique_col_values(df)
```

```python
df = df.replace("No internet service","No")
df = df.replace("No phone service","No")

print_unique_col_values(df)
```

```python
cols_to_scale = ['tenure','MonthlyCharges','TotalCharges']

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df[cols_to_scale] = scaler.fit_transform(df[cols_to_scale])
```

```python
mapping = {'No': 0, 'Yes': 1} 
df['Churn'] = df['Churn'].map(mapping)
```

```python
df.info()
```

# **Model and Prediction part: -**

```python
x = df.drop('Churn',axis = 'columns')
y = df['Churn']

from sklearn.model_selection import train_test_split
x_train,x_test,y_train ,y_test = train_test_split(x,y,test_size = 0.2,random_state=42)
```

```python
x_train.shape
```

```python
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### **Build a model: -**

```python
import tensorflow as tf
from tensorflow import keras

model = keras.Sequential([
    keras.layers.Dense(23,input_shape =(23,) ,activation = 'relu'),
    keras.layers.Dense(15,activation = 'relu'),
    keras.layers.Dense(10,activation = 'relu'),
    # keras.layers.Dense(6,activation = 'relu'),
    keras.layers.Dense(1,activation = 'sigmoid')
])
```

```python
model.compile(optimizer = 'adam',
        loss = 'binary_crossentropy',
        metrics = ['accuracy']
)
model.fit(x_train,y_train,epochs=100)
```

```python
yp = model.predict(x_test)
yp[:5]
```

```python
y_pred = []
for element in yp:
    if element > 0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
y_pred[:10]
```

```python
y_test[:10]
```

```python
from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(y_test,y_pred))
```

```python
import seaborn as sn
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_pred)

plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
