### **Simple Preprocess and Tokenization: -**

```python
df1 = df1.apply(gensim.utils.simple_preprocess)
df1.head()
```

```python
print(k[0])
print(df1[0])
```

### **Train a Word2Vec model**

```python
import os
os.cpu_count()
```

```python
import multiprocessing

model = gensim.models.Word2Vec(
    window = 5,
    min_count= 2,
    workers = multiprocessing.cpu_count() - 1,
)
```

### Build a vocabulary: -

```python
model.build_vocab(df1,progress_per=1000)
```

```python
model.corpus_count
```

```python
model.epochs
```

```python
model.train(df1,total_examples=model.corpus_count,epochs= model.epochs)
```

### **Model saved: -**

```python
model.save("./word2vec-amazon-cell-accessories-reviews-short.model")
```

### **Find Similarity between Words: -**

```python
model.wv.most_similar("bad")
```

```python
model.wv.similarity(w1="great", w2="lower")
```
