```python
# Pandas using with jason :-)
import pandas as pd
from io import StringIO
```

```python
data = '{"Employee_name" : "Harry","Email" : "harry@gmail.com", "Job_profile" : [{"Title1" : "Ethecial_Hacker", "Tile2" : "Ai Engineer"}]}'
```

```python
# Creating a dataframe from string in jason foramt :-
df = pd.read_json(StringIO(data))
df
```

```python
type(df)
```

```python
pd.read_json(StringIO(data),orient = 'record')
```

```python
pd.read_json(StringIO(data),orient = 'index')
```

```python
# pd.read_json(StringIO(data),orient = 'split')
pd.read_json(StringIO(data),orient = 'column')
```

```python
# Cinverting dataframe into json :-
df = pd.DataFrame(data = [['a','b'],['c','d']],index = ["Row1","Row2"],columns =["Column1","Column2"] )
type(df)
df
```

```python

df.to_json()      # default it convert into column format :-
df.to_json(orient = 'columns')
```

```python
df.to_json(orient = 'index')    # convert into row format:-
```

```python
df.to_json(orient  = 'split')  # Seperate everything :-
```

```python
df.to_json(orient = 'table')     # give all details :-
```
