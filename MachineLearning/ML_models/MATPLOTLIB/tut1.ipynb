```python
# Creating a Legend in graph :-).

fig = plt.figure()
v = plt.subplot()
plt.title("The Great Ninja War Performance Graph :")
v.plot("f1","f0",'b:o',label = "Good",data = A)
v.plot("f3","f1",'k--*',label= "Bad",data = A)

l = v.legend()
# v.legend(loc = 'upper center')
# v.legend(loc = 'upper right')
# v.legend(loc = 'upper left')
# v.legend(loc = 'center')
# v.legend(loc = 'lower center')
# v.legend(loc = 'lower right')
# v.legend(loc = 'lower left') 
```

```python
# changing the background colour of legend : )

l.get_frame().set_facecolor('grey')
fig
```

```python
# Adjust the font size of legend

v.legend(loc = 'upper center', fontsize = 'xx-large')
v.legend(loc = 'upper center', fontsize = 'x-large')
v.legend(loc = 'upper center', fontsize = 'large')
v.legend(loc = 'upper center', fontsize = 'small')
v.legend(loc = 'upper center', fontsize = 'x-small')
v.legend(loc = 'upper center', fontsize = 'xx-small')
fig
```
