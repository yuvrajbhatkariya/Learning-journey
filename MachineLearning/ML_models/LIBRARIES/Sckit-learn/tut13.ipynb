## **K-cross Validation: -**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
```

```python
lr = LogisticRegression()
svm = SVC()
dt = tree.DecisionTreeClassifier(random_state = 42)
rf = RandomForestClassifier(n_estimators = 40,random_state = 42)
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
dir(iris)
```

```python
df = pd.DataFrame(iris.data)
df['target'] = iris.target
df.head()
```

```python
X = iris.data
Y = iris.target
```

## **Use train_test_split : -**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.2)
```

```python
lr.fit(x_train,y_train)
y_pred = lr.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
svm.fit(x_train,y_train)
y_pred = svm.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
dt.fit(x_train,y_train)
y_pred = dt.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
rf.fit(x_train,y_train)
y_pred = rf.predict(x_test)
accuracy_score(y_test,y_pred)
```

## **Use k - fold cross validation: -**

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
kf = KFold(n_splits = 5)
kf
```

```python
print("Logistic Regression : ",cross_val_score(lr,X,Y))
print("SVM : ",cross_val_score(svm,X,Y))
print("Decision Tree : ",cross_val_score(dt,X,Y))
print("Random Forest : ",cross_val_score(rf,X,Y))
```
