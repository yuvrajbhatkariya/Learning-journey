## **Apply K-Means cluster: -**

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters = 3)
km
```

```python
y_pred = km.fit_predict(df[[0,1]])
y_pred
```

```python
df['Cluster'] = y_pred
df.head()
```

```python
df1 = df[df.Cluster == 0]
df2 = df[df.Cluster == 1]
df3 = df[df.Cluster == 2]
```

```python
km.cluster_centers_
```

```python
plt.scatter(df1[0],df1[1],color = 'green',marker = '.')
plt.scatter(df2[0],df2[1],color = 'blue',marker = '.')
plt.scatter(df3[0],df3[1],color = 'red',marker = '.')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color = 'black',marker = '*')
```
