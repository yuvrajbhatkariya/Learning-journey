```python
import requests

url = "http://localhost:11434/api/generate"

while True:
    user = input("You: ")
    if user.lower() == "exit":
        break

    payload = {
        "model": "mistral",
        "prompt": user,
        "stream": False
    }

    res = requests.post(url, json=payload)
    print("AI:", res.json()["response"])
```
