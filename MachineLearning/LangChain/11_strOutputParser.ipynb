```python
from langchain_huggingface import ChatHuggingFace ,HuggingFaceEndpoint
from langchain_core.prompts import PromptTemplate
from dotenv import load_dotenv

load_dotenv()
```

```python
llm = HuggingFaceEndpoint(
    repo_id='HuggingFaceH4/zephyr-7b-beta',
    task='text-generation',
    max_new_tokens=200,
    temperature=0.2
)
model = ChatHuggingFace(llm=llm)
```

```python
from langchain_core.messages import HumanMessage

result = model.invoke([
    HumanMessage(content="How many players are in a football team?")
])

print(result.content)
```

```python
template1 = PromptTemplate(
    template="write a detetail report about {topic}",
    input_variables=["topic"]
)
template2 = PromptTemplate(
    template="give ahort 4-5 line summary on the folowing text. \n {text}",
    input_variables=["text"]
)
```

```python
prompt1 = template1.invoke({'topic' : 'Transformer'})

result = model.invoke(prompt1)

prompt2 = template2.invoke({'text' : result.content})

result = model.invoke(prompt2)

print(result.content)

```

### **Now we use output parser : -**

```python
from langchain_core.output_parsers import StrOutputParser

parser = StrOutputParser()

chain = template1 | model | parser | template2 | model | parser

result = chain.invoke({"topic" : "black Panther"})

print(result)
```
